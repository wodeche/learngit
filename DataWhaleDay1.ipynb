## 请求头


请求头就是告诉被请求的服务器需要传送什么样格式的信息。

    -Accept：浏览器可接受的MIME类型
    -Accept-Charset:浏览器可接受的字符集
    -Accept-Language:浏览器所希望的语言种类
    -Authorization：授权信息
    -Connection：表示是否需要持久连接
    -Content-Length:表示消息正文长度
    -Cookie：最重要的请求头之一
    -User-Agent:浏览器类型

## 如何添加请求头

```python
headers={"User-Agent" : "Mozilla/5.0 (Windows; U; Windows NT 5.1; zh-CN; rv:1.9.1.6) ",
  "Accept" : "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
  "Accept-Language" : "en-us",
  "Connection" : "keep-alive",
  "Accept-Charset" : "GB2312,utf-8;q=0.7,*;q=0.7"}

r=requests.post("http://baike.baidu.com/item/哆啦A梦",headers=headers,allow_redirects=False)   #allow_redirects设置为重定向

r.encoding='UTF-8'
print(r.url)
print(r.headers) #响应头
print(r.request.headers) #请求头
```

## python正则表达式

re模块使python拥有全部正则表达式的功能

### re.match函数

re.match尝试从起始位置匹配一个模式re.match(pattern,string,flags=0)


    1.pattern:正则表达式
    2.string：匹配字符串
    3.flags：正则表达式匹配方式

### re.search函数

re.search函数扫描整个字符串，并返回第一个成功的匹配。re.search(pattern,string,flags=0)

### re.match和re.search的区别

re.match值匹配字符串的开始，如果开始不符合则匹配失败，返回None；
而re.search匹配整个字符串，直到找到一个匹配

### 检索与替换

re.sub用于替换字符串的匹配项。re.sub(pattern,repl,string,count =0,flags=0)

    1.pattern:正则模式字符串
    2.repl：替换字符串，也可以是函数
    3.string：原始字符串
    4.count：模式匹配替换的最大次数，0代表替换所有

### re.compile函数

compile函数用于编译正则表达式，生成一个pattern对象

语法为：re.compile(pattern[,flags])

    1pattern:一个字符串形式正则
    2：flags：匹配模式

### re.findall函数

在字符串中找到正则表达式所匹配的所有子串，并返回一个列表：re.findall(string,pos,endpos)

### 抓取豆瓣TOP250

```python
import requests
import re
import csv


def movie_info(url):
    headers = {
     'User-Agent':"Mozilla/5.0 (Macintosh; U; Intel Mac OS X 10_6_8; en-us) AppleWebKit/534.50 (KHTML, like Gecko) Version/5.1 Safari/534.50"
    }

    res = requests.get(url, headers=headers)
    ranks = re.findall(' <em class="">(.*?)</em>',res.text, re.S)
    names = re.findall('<span class="title">([\u4e00-\u9fa5]+)</span>',res.text, re.S)
    countries = re.findall('&nbsp;/&nbsp;([\u4e00-\u9fa5]+)&nbsp;/&nbsp;', res.text, re.S)
    text = re.sub('导演: ',"",res.text)  
    directors = re.findall('<p class="">(.*?)&nbsp;&nbsp;', text, re.S)
    scores = re.findall('<span class="rating_num" property="v:average">(.*?)</span>',res.text,re.S)

    for rank,name,country,director,score in zip(ranks,names,countries,directors,scores):
        writer.writerow([rank,name,country,director,score])


if __name__ == '__main__':

    file = open('E:/web crawler/movie.csv','w+',encoding='utf-8',newline='')
    writer = csv.writer(file)
    writer.writerow(['rank','name','country','director','score'])

    for i in range(0,250,25):
        url = 'https://movie.douban.com/top250?start={}&filter='.format(i)
        movie_info(url)
```

```python
#!/usr/bin/env python 
# -*- coding:utf-8 -*-
import requests
import re
import json
from requests.exceptions import RequestException
#得到网页源代码
def get_one_page(url):
    try:
        res=requests.get(url)
        if res.status_code==200:
            return  res.text
        return  None
    except RequestException:
        return  None
#对源代码进行解析将要获取的数据存储到dict当中去然后写入本地文件
def parse_one_html(html):
    regex='<em class="">(\d+)</em>.*?<span class="title">(.*?)</span>.*?<p class="">(.*?)</p>.*?<span class="rating_num" property="v:average">(.*?)</span>.*?<span>(.*?)</span>.*?<span class="inq">(.*?)</span>'
    pattern=re.compile(regex,re.S)
    items=re.findall(pattern,html)
    #对得到的list列表进行处理，将每条记录变成一个dict
    for item in items:
        content=""
        for every_list in item[2].split():
            content=content+"".join(every_list)
        content=re.sub('&nbsp;',' ',content)
        content=re.sub('<br>', '', content)
        #将获取到的信息存储到dict字典当中
        dict={
            "number":item[0],
            "name":item[1],
            "describe":content,
            "star":item[3],
            "evaluate":item[4],
            "title":item[5]
 
        }
        #将得到的dict写入本地文件当中去
        with open('result.txt','a',encoding='utf-8') as f:
            f.write(json.dumps(dict,ensure_ascii=False)+'\n')
def main(start):
    url='https://movie.douban.com/top250'
    if start!=0:
        url='https://movie.douban.com/top250?start='+str(start)+'&filter='
    html=get_one_page(url)
    parse_one_html(html)
if __name__=='__main__':
    for i in range(10):
        main(i*25)
```
